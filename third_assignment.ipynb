```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load dataset
data = pd.read_csv('Housing.csv')

# Display the first few rows of the dataset
data.head()
```

```python
# Print all column names to debug KeyError issues
print("Columns in DataFrame:", data.columns.tolist())

# Strip whitespace from headers (if any)
data.columns = data.columns.str.strip()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='bedrooms', y='price', data=data, estimator=lambda x: sum(x) / len(x))
plt.title('Average Price by Number of Bedrooms')
plt.xlabel('Number of Bedrooms')
plt.ylabel('Average Price')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data['price'], bins=30, kde=True)
plt.title('Distribution of Property Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

```python
# Bar chart for average price by number of bedrooms
plt.figure(figsize=(10, 6))
sns.barplot(x='bedrooms', y='price', data=data, estimator=lambda x: sum(x) / len(x))
plt.title('Average Price by Number of Bedrooms')
plt.xlabel('Number of Bedrooms')
plt.ylabel('Average Price')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='stories', y='price', data=data)
plt.title('Price Distribution by Number of Stories')
plt.xlabel('Number of Stories')
plt.ylabel('Price')
plt.show()
```
